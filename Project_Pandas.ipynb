Task 32: Create a column "RevenueCategory" with three levels:

"Low" for transactions under 10 "Medium" for transactions between 10 and 100 "High" for transactions above 100

```python
def revenue_category(x):
    if x < 10:
        return "Low"
    elif 10 <= x <= 100:
        return "Medium"
    else:
        return "High"

odf['RevenueCategory'] = odf['TotalPrice'].apply(revenue_category)
```

```python
odf
```

Task 33: Create a column "OrderSize" with three levels:

"Small" for Quantity less than 5 "Medium" for Quantity between 5 and 20 "Large" for Quantity greater than 20

```python
def Ordersize(a):
  if a <5:
    return "Small"
  elif a >5 and a <20:
    return "Medium"
  else:
    return "Large"

odf['OrderSize']=odf['Quantity'].apply(Ordersize)
```
